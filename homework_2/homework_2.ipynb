```python
# importing pandas lib
import pandas as pd
# importing file for 1st task
db_1 = pd.read_csv('teams_stats.csv')
# clear nan values in required columns
db_1[['Year','Tm']].dropna(inplace=True)
# creating crosstable with Years as rows and Teams as columns, where values are frequencies of combination of year and row
db_1 = pd.crosstab(db_1['Year'], db_1['Tm'])
print(db_1)
```

```python
# if frequency is larger than 0 in each row, than we can count it as "unique team in particular year"
db_1['Unique_teams_count'] = db_1[db_1>0].count(axis=1)
db_1['Unique_teams_count']
```
